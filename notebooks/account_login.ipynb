```python
t_cols = ['log_time', 'player_id', 'country', 'lv', 'os', 'device', 'yc', 'cheer_point', 'yp', 'sp', 'ep', 'type', 'is_daily_first_login']
df = pd.read_csv('data/csv_log_account_login.csv', names=t_cols)
```

```python
import time
def chg_time(log_time):
    return time.ctime(log_time)
```

```python
df['log_time'] = df['log_time'].apply(chg_time)
```

```python
df.index = pd.to_datetime(df.pop('log_time'))
```

```python
# jsob format
#] sdate='2014-10-10' edate='2014-10-31' runipy --stdout account_login.ipynb
from os import environ
sdate = environ.get('sdate', '2014-10-01')
edate = environ.get('edate', '2014-10-31')
print sdate + " ~ " + edate
```

```python
grouped_player = df[sdate:edate].groupby('player_id').size()
grouped_player.sort()
grouped_player.plot(kind='barh', stacked=True, figsize=(18, 8))
```

```python
t_span = '1D'
t_player_id = grouped_player[-5:].index
player_data = {}
for no in t_player_id:
    player_data[no] = df['player_id'][df['player_id'] == no].resample(t_span, how='count')
df_player = DataFrame(player_data)
df_player.head()
```

```python
df_player.plot(figsize=(18, 8))
```

```python
t_span = '1D'
t_player_id = grouped_player[-5:].index
level_data = {}
for no in t_player_id:
    level_data[no] = df['lv'][df['player_id'] == no].resample(t_span, how='max')
df_level = DataFrame(level_data)
df_level.head()
```

```python
df_level.plot(figsize=(18, 8))
```

```python
grouped_device = df.groupby('device').size()
grouped_device.sort()
device_plot = grouped_device.plot(kind='barh', stacked=True, figsize=(18, 8))
```
