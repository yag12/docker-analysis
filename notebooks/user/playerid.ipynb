```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame, Panel, concat
from os import environ, path
import time
import glob
```

```python
sdate = environ.get('sdate', '2014-10-01')
edate = environ.get('edate', '2014-10-31')
player_id = environ.get('player_id', 14)
isdate = int(sdate.replace("-", "") + "0000")
iedate = int(edate.replace("-", "") + "2359")
player_id = int(player_id)
```

```python
def chg_date(log_time):
    return time.strftime("%Y-%m-%d", time.localtime(log_time))

def chg_time(log_time):
    return time.strftime("%H", time.localtime(log_time))

def chg_timestamp(log_time):
    return time.ctime(log_time)
```

```python
t_cols = ['log_time', 'player_id', 'country', 'lv', 'os', 'device', 'yc', 'cheer_point', 'yp', 'sp', 'ep', 'type', 'is_daily_first_login']
log_file = glob.glob('/logfile/account/*_csv_log_account_login.csv')
df = None

for fname in log_file:
    regdate = int(path.basename(fname).split("_")[4])
    if isdate <= regdate and iedate >= regdate:
        csv = pd.read_csv(fname, names=t_cols)
        if df is not None:
            df = concat([df[0:], csv[0:]])
        else:
            df = csv
```

```python
df['player_id'] = df['player_id'].astype('int')
df = df[df['player_id'] == player_id]
```

```python
df_data = DataFrame(df[0:])
df_data['log_time'] = df_data['log_time'].apply(chg_timestamp)
df_data.index = pd.to_datetime(df_data.pop('log_time'))
df_data.sort()
```

```python
df_time = DataFrame(df[0:])
df_time['time'] = df_time['log_time'].apply(chg_time)
df_time = DataFrame({"Access":df_time.groupby('time').size(), "First":df_time[df_time['is_daily_first_login'] == 1].groupby('time').size()})
df_time[np.isnan(df_time)] = 0
df_time.plot()
```

```python
t_span = '1D'
df_data = df_data.rename(columns={'log_time':'day'})
df_player = DataFrame({"Access":df_data['player_id'].resample(t_span, how='count')})
df_player.plot()
```
